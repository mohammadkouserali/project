```python
# ============================================================
# CELL 1: INSTALLS + IMPORTS
# ============================================================
import os
import sys
import gc
import cv2
import time
import glob
import shutil
import random
import zipfile
import warnings
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from pathlib import Path
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_curve,
    auc,
    roc_auc_score,
)
from sklearn.utils.class_weight import compute_class_weight
from google.colab import drive
from IPython.display import display, HTML

from tensorflow.keras import layers, models, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint

warnings.filterwarnings("ignore")

try:
    import cv2  # noqa
except ImportError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "opencv-python-headless"])
    import cv2

print("TensorFlow version:", tf.__version__)


# ============================================================
# CELL 2: GOOGLE DRIVE + CONFIG
# ============================================================
drive.mount("/content/drive", force_remount=False)

# -----------------------------
# DATA PATHS
# -----------------------------
RAW_ZIP_FILE_PATH = "/content/drive/MyDrive/1/isic2020.zip"

PROJECT_ROOT    = "/content/melanoma_project_raw_fair"
RAW_EXTRACT_DIR = os.path.join(PROJECT_ROOT, "raw_extracted")
OUTPUT_DIR      = os.path.join(PROJECT_ROOT, "outputs")

# -----------------------------
# GENERAL SETTINGS
# -----------------------------
SEED = 42

CLASSIFIER_INPUT_SHAPE = (224, 224, 3)
CLASSIFIER_IMG_SIZE    = CLASSIFIER_INPUT_SHAPE[:2]

DEFAULT_BATCH_SIZE  = 32

PHASE1_EPOCHS          = 15
PHASE2_EPOCHS          = 35
HEAD_LEARNING_RATE     = 1e-3
FINETUNE_LEARNING_RATE = 1e-5
LABEL_SMOOTHING        = 0.0
USE_CLASS_WEIGHTS      = True
FINE_TUNE_FRACTION     = 0.30

TTA_STEPS = 5

# -----------------------------
# BACKBONES TO RUN (Two models only)
# -----------------------------
BACKBONE_NAMES = ["MobileNetV2_S", "EfficientNetB3"]

# -----------------------------
# THRESHOLD TUNING
# -----------------------------
THRESHOLD_MIN = 0.10
THRESHOLD_MAX = 0.90
THRESHOLD_STEPS = 161
MIN_SPECIFICITY_FOR_THRESHOLD = 0.95

for p in [PROJECT_ROOT, RAW_EXTRACT_DIR, OUTPUT_DIR]:
    os.makedirs(p, exist_ok=True)

print("Project root :", PROJECT_ROOT)
print("Backbones    :", BACKBONE_NAMES)
print("Output dir   :", OUTPUT_DIR)


# ============================================================
# CELL 3: GENERAL UTILITIES
# ============================================================
def set_global_seed(seed=SEED):
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)

def ensure_dir(path):
    os.makedirs(path, exist_ok=True)

def unzip_dataset(zip_path, extract_to):
    if not os.path.exists(zip_path):
        raise FileNotFoundError(f"ZIP not found: {zip_path}")

    ensure_dir(extract_to)

    if len(os.listdir(extract_to)) > 0:
        print(f"Already extracted: {extract_to}")
        return True

    print(f"Extracting: {zip_path}")
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(extract_to)
    print(f"Extracted to: {extract_to}")
    return True

def find_base_dir(root_dir):
    if os.path.exists(os.path.join(root_dir, "train")):
        return root_dir
    for r, dirs, _ in os.walk(root_dir):
        if {"train", "val", "test"}.issubset(set(dirs)):
            return r
    return root_dir

def infer_positive_class(class_indices):
    inverse = {v: k for k, v in class_indices.items()}
    for idx, name in inverse.items():
        if "mel" in name.lower():
            return idx, name
    max_idx = max(inverse.keys())
    return max_idx, inverse[max_idx]

def get_negative_class_name(class_indices, positive_idx):
    inverse = {v: k for k, v in class_indices.items()}
    for idx, name in inverse.items():
        if idx != positive_idx:
            return name
    return "Negative"

set_global_seed(SEED)
print("Utilities ready.")


# ============================================================
# CELL 4: BACKBONE FACTORY + DATA GENERATORS
# ============================================================
def get_backbone(backbone_name):
    if backbone_name == "MobileNetV2_S":
        from tensorflow.keras.applications import MobileNetV2
        from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

        def fn(include_top=False, weights="imagenet", input_tensor=None, pooling="avg"):
            return MobileNetV2(
                include_top=include_top,
                weights=weights,
                input_tensor=input_tensor,
                pooling=pooling,
                alpha=0.75
            )
        return fn, preprocess_input, "lightweight"

    if backbone_name == "EfficientNetB3":
        from tensorflow.keras.applications import EfficientNetB3
        from tensorflow.keras.applications.efficientnet import preprocess_input

        def fn(include_top=False, weights="imagenet", input_tensor=None, pooling="avg"):
            return EfficientNetB3(
                include_top=include_top,
                weights=weights,
                input_tensor=input_tensor,
                pooling=pooling
            )
        return fn, preprocess_input, "stronger"

    raise ValueError(f"Unknown backbone: {backbone_name}")

def make_generators(base_dir, backbone_name):
    _, preprocess_input, _ = get_backbone(backbone_name)
    bs = DEFAULT_BATCH_SIZE

    def prep(img):
        return preprocess_input(img.astype(np.float32))

    train_gen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.03,
        height_shift_range=0.03,
        zoom_range=0.08,
        horizontal_flip=True,
        fill_mode="nearest",
        preprocessing_function=prep,
    ).flow_from_directory(
        os.path.join(base_dir, "train"),
        target_size=CLASSIFIER_IMG_SIZE,
        batch_size=bs,
        class_mode="binary",
        interpolation="bilinear",
        shuffle=True,
        seed=SEED,
    )

    val_test_dg = ImageDataGenerator(preprocessing_function=prep)

    val_gen = val_test_dg.flow_from_directory(
        os.path.join(base_dir, "val"),
        target_size=CLASSIFIER_IMG_SIZE,
        batch_size=bs,
        class_mode="binary",
        interpolation="bilinear",
        shuffle=False,
    )

    test_gen = val_test_dg.flow_from_directory(
        os.path.join(base_dir, "test"),
        target_size=CLASSIFIER_IMG_SIZE,
        batch_size=1,
        class_mode="binary",
        interpolation="bilinear",
        shuffle=False,
    )

    return train_gen, val_gen, test_gen

def get_class_weights(train_gen):
    y = train_gen.classes
    cw = compute_class_weight("balanced", classes=np.unique(y), y=y)
    d = {int(c): float(w) for c, w in zip(np.unique(y), cw)}
    print("Class weights:", d)
    return d


# ============================================================
# TTA PREDICTION FUNCTION
# ============================================================
def predict_with_tta(model, test_dir, backbone_name, tta_steps=TTA_STEPS):
    """
    Runs inference tta_steps times on the test set:
      - 1 pass with no augmentation (original)
      - (tta_steps - 1) passes with horizontal flip + small rotation + zoom
    Returns averaged probabilities, ground-truth labels, and class_indices.
    """
    _, preprocess_input, _ = get_backbone(backbone_name)

    def prep(img):
        return preprocess_input(img.astype(np.float32))

    # Pass 1: no augmentation
    base_dg = ImageDataGenerator(preprocessing_function=prep)
    base_gen = base_dg.flow_from_directory(
        test_dir,
        target_size=CLASSIFIER_IMG_SIZE,
        batch_size=1,
        class_mode="binary",
        interpolation="bilinear",
        shuffle=False,
    )
    print(f"TTA pass 1/{tta_steps} (original) ...")
    all_preds = model.predict(base_gen, steps=len(base_gen), verbose=1)

    # Passes 2..N: with augmentation
    tta_dg = ImageDataGenerator(
        horizontal_flip=True,
        rotation_range=10,
        zoom_range=0.05,
        fill_mode="nearest",
        preprocessing_function=prep,
    )

    for step in range(2, tta_steps + 1):
        tta_gen = tta_dg.flow_from_directory(
            test_dir,
            target_size=CLASSIFIER_IMG_SIZE,
            batch_size=1,
            class_mode="binary",
            interpolation="bilinear",
            shuffle=False,
        )
        print(f"TTA pass {step}/{tta_steps} (augmented) ...")
        all_preds += model.predict(tta_gen, steps=len(tta_gen), verbose=1)

    avg_preds = (all_preds / tta_steps).ravel()
    return avg_preds, base_gen.classes.astype(int), base_gen.class_indices


# ============================================================
# CELL 5: MODEL + TRAINING UTILITIES
# ============================================================
def build_model(backbone_name):
    bb_fn, _, profile = get_backbone(backbone_name)

    print("=" * 70)
    print(f"BUILDING: {backbone_name}  [{profile}]")
    print("=" * 70)

    inp = Input(shape=CLASSIFIER_INPUT_SHAPE, name="input_image")
    base = bb_fn(include_top=False, weights="imagenet", input_tensor=inp, pooling="avg")
    base.trainable = False

    x = layers.BatchNormalization(name="head_bn0")(base.output)
    x = layers.Dropout(0.35, name="head_do0")(x)

    x = layers.Dense(512, name="head_fc1")(x)
    x = layers.BatchNormalization(name="head_fc1_bn")(x)
    x = layers.LeakyReLU(negative_slope=0.2, name="head_fc1_act")(x)
    x = layers.Dropout(0.40, name="head_do1")(x)

    x = layers.Dense(256, name="head_fc2")(x)
    x = layers.BatchNormalization(name="head_fc2_bn")(x)
    x = layers.LeakyReLU(negative_slope=0.2, name="head_fc2_act")(x)
    x = layers.Dropout(0.30, name="head_do2")(x)

    out = layers.Dense(1, activation="sigmoid", name="classifier")(x)

    model = models.Model(inputs=inp, outputs=out, name=f"{backbone_name}_Transfer")

    print(f"Total params:    {model.count_params():,}")
    print(f"Backbone params: {base.count_params():,}")
    return model, base

def compile_model(model, lr):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss=tf.keras.losses.BinaryCrossentropy(label_smoothing=LABEL_SMOOTHING),
        metrics=[
            tf.keras.metrics.BinaryAccuracy(name="accuracy"),
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc"),
        ],
    )

def unfreeze_top(base, fraction=FINE_TUNE_FRACTION):
    base.trainable = True
    n = len(base.layers)
    cut = int(n * (1.0 - fraction))

    for i, layer in enumerate(base.layers):
        layer.trainable = i >= cut and not isinstance(layer, layers.BatchNormalization)

    print(f"Fine-tuning top {fraction * 100:.0f}% ({n - cut}/{n} layers)")

def make_callbacks(run_name, run_dir, phase):
    ckpt = os.path.join(run_dir, f"best_{run_name}_{phase}.keras")
    return [
        ModelCheckpoint(ckpt, monitor="val_auc", mode="max", save_best_only=True, verbose=1),
        ReduceLROnPlateau(
            monitor="val_auc",
            mode="max",
            factor=0.5,
            patience=4,
            min_lr=1e-7,
            verbose=1
        ),
        EarlyStopping(
            monitor="val_auc",
            mode="max",
            patience=8,
            min_delta=1e-4,
            restore_best_weights=True,
            verbose=1
        ),
    ], ckpt

def merge_histories(h1, h2):
    keys = sorted(set(list(h1.history) + list(h2.history)))
    return {k: h1.history.get(k, []) + h2.history.get(k, []) for k in keys}

def plot_history(train_vals, val_vals, title, ylabel, save_path):
    if not train_vals and not val_vals:
        return

    plt.figure(figsize=(8, 6))
    if train_vals:
        plt.plot(range(1, len(train_vals) + 1), train_vals, label=f"Train {ylabel}", linewidth=2)
    if val_vals:
        plt.plot(range(1, len(val_vals) + 1), val_vals, label=f"Val {ylabel}", linewidth=2)

    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()


# ============================================================
# CELL 6: THRESHOLD TUNING + EVALUATION
# ============================================================
def plot_cm(cm, labels, title, path):
    plt.figure(figsize=(8, 6))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=labels,
        yticklabels=labels,
        cbar=False,
        annot_kws={"size": 14},
    )
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()

def plot_roc(y_true, y_score, title, path, pos_name="Positive"):
    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, lw=2, label=f"{pos_name} (AUC = {roc_auc:.4f})")
    plt.plot([0, 1], [0, 1], "--", lw=1)
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.2)
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()

def choose_best_threshold(y_true, y_prob, min_specificity=0.95):
    best_t = 0.5
    best_score = -1.0
    rows = []

    thresholds = np.linspace(THRESHOLD_MIN, THRESHOLD_MAX, THRESHOLD_STEPS)

    for t in thresholds:
        y_pred = (y_prob >= t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred, labels=[0, 1]).ravel()

        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0.0
        recall      = recall_score(y_true, y_pred, zero_division=0)
        precision   = precision_score(y_true, y_pred, zero_division=0)
        f1          = f1_score(y_true, y_pred, zero_division=0)

        if min_specificity is not None and specificity < min_specificity:
            continue

        rows.append({
            "threshold":   t,
            "specificity": specificity,
            "recall":      recall,
            "precision":   precision,
            "f1":          f1
        })

        if f1 > best_score:
            best_score = f1
            best_t = t

    df = pd.DataFrame(rows)

    if df.empty:
        print("No threshold satisfied the specificity constraint. Falling back to 0.5")
        return 0.5, df

    return float(best_t), df

def normalize_positive_class(y_true, y_prob, class_indices):
    pos_idx, pos_name = infer_positive_class(class_indices)
    neg_name = get_negative_class_name(class_indices, pos_idx)

    y_true = y_true.astype(int).copy()
    y_prob  = y_prob.copy()

    if pos_idx != 1:
        y_true = 1 - y_true
        y_prob  = 1.0 - y_prob

    return y_true, y_prob, pos_name, neg_name

def evaluate_model(model, val_gen, test_dir, backbone_name, run_name, run_dir):
    print("=" * 70)
    print(f"EVALUATION: {run_name}")
    print("=" * 70)

    # --- Validation probs for threshold tuning (no TTA on val) ---
    val_probs = model.predict(val_gen, steps=len(val_gen), verbose=1).ravel()
    y_val     = val_gen.classes.astype(int)

    y_val, val_probs, pos_name, neg_name = normalize_positive_class(
        y_val, val_probs, val_gen.class_indices
    )

    best_thresh, thr_df = choose_best_threshold(
        y_val, val_probs, min_specificity=MIN_SPECIFICITY_FOR_THRESHOLD
    )

    thr_csv = os.path.join(run_dir, f"threshold_search_{run_name}.csv")
    thr_df.to_csv(thr_csv, index=False)
    print(f"Chosen threshold: {best_thresh:.4f}")
    print("Saved threshold table:", thr_csv)

    # --- Test probs with TTA ---
    t0 = time.time()
    probs, y_true, class_indices = predict_with_tta(model, test_dir, backbone_name, TTA_STEPS)
    elapsed = time.time() - t0
    per_img = elapsed / max(len(probs), 1)

    y_true, probs, pos_name, neg_name = normalize_positive_class(
        y_true, probs, class_indices
    )

    y_pred = (probs >= best_thresh).astype(int)

    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()

    specificity = (tn / (tn + fp) * 100) if (tn + fp) > 0 else 0.0
    roc_auc     = roc_auc_score(y_true, probs)

    plot_roc(
        y_true, probs,
        f"ROC Curve - {run_name}",
        os.path.join(run_dir, f"roc_{run_name}.png"),
        pos_name=pos_name
    )

    plot_cm(
        cm,
        [neg_name, pos_name],
        f"Confusion Matrix - {run_name}",
        os.path.join(run_dir, f"cm_{run_name}.png")
    )

    np.savez(
        os.path.join(run_dir, f"predictions_{run_name}.npz"),
        y_true=y_true, y_pred=y_pred, y_prob=probs, threshold=best_thresh,
    )

    acc  = accuracy_score(y_true, y_pred) * 100
    pre  = precision_score(y_true, y_pred, average="weighted", zero_division=0) * 100
    rec  = recall_score(y_true, y_pred, average="weighted", zero_division=0) * 100
    f1w  = f1_score(y_true, y_pred, average="weighted", zero_division=0) * 100

    p_pr = precision_score(y_true, y_pred, pos_label=1, average="binary", zero_division=0) * 100
    p_re = recall_score(y_true, y_pred, pos_label=1, average="binary", zero_division=0) * 100
    p_f1 = f1_score(y_true, y_pred, pos_label=1, average="binary", zero_division=0) * 100

    mdf = pd.DataFrame({
        "Metric": [
            "Chosen Threshold", "TTA Steps",
            "Accuracy (ACC)", "Precision Weighted", "Recall Weighted",
            "Specificity (SPE)", "F1 Weighted",
            f"{pos_name} Precision", f"{pos_name} Recall", f"{pos_name} F1",
            "ROC AUC", "Test Set Time", "Per Image Time",
        ],
        "Value": [
            f"{best_thresh:.4f}", f"{TTA_STEPS}",
            f"{acc:.2f}%", f"{pre:.2f}%", f"{rec:.2f}%",
            f"{specificity:.2f}%", f"{f1w:.2f}%",
            f"{p_pr:.2f}%", f"{p_re:.2f}%", f"{p_f1:.2f}%",
            f"{roc_auc:.4f}", f"{elapsed:.2f} s", f"{per_img:.4f} s",
        ],
    })

    csv_path = os.path.join(run_dir, f"metrics_{run_name}.csv")
    mdf.to_csv(csv_path, index=False)
    print("Saved:", csv_path)

    try:
        display(HTML(mdf.to_html(index=False)))
    except Exception:
        print(mdf.to_string(index=False))

    return {
        "Run": run_name, "Backbone": backbone_name, "TTA Steps": TTA_STEPS,
        "ACC": acc, "PRE": pre, "REC": rec, "SPE": specificity, "F1 Score": f1w,
        f"{pos_name} Precision": p_pr, f"{pos_name} Recall": p_re, f"{pos_name} F1": p_f1,
        "ROC AUC": roc_auc, "Threshold": best_thresh,
        "Execution Time (s)": elapsed, "Per Image Test Time (s)": per_img,
    }


# ============================================================
# CELL 7: PREPARE RAW DATASET
# ============================================================
def prepare_raw_dataset():
    ensure_dir(PROJECT_ROOT)
    ensure_dir(OUTPUT_DIR)

    unzip_dataset(RAW_ZIP_FILE_PATH, RAW_EXTRACT_DIR)

    raw_base = find_base_dir(RAW_EXTRACT_DIR)
    print("Raw base:", raw_base)

    if not os.path.exists(os.path.join(raw_base, "train")):
        raise FileNotFoundError("train/val/test not found in raw dataset.")

    return raw_base


# ============================================================
# CELL 8: RUN EXPERIMENT FOR ONE BACKBONE
# ============================================================
def run_raw_experiment(backbone_name, base_dir):
    tf.keras.backend.clear_session()
    gc.collect()
    set_global_seed(SEED)

    safe     = "raw_fair"
    run_name = f"{backbone_name}_{safe}"
    run_dir  = os.path.join(OUTPUT_DIR, backbone_name, safe)
    ensure_dir(run_dir)

    test_dir = os.path.join(base_dir, "test")

    print("\n" + "#" * 90)
    print(f"  RUNNING: {run_name}")
    print("#" * 90)
    print("Data dir:", base_dir)

    train_gen, val_gen, test_gen = make_generators(base_dir, backbone_name)

    print(f"Train: {train_gen.samples}  Val: {val_gen.samples}  Test: {test_gen.samples}")
    print("Classes:", train_gen.class_indices)
    print(f"TTA steps: {TTA_STEPS}")

    cw = get_class_weights(train_gen) if USE_CLASS_WEIGHTS else None

    model, base = build_model(backbone_name)

    # ----- PHASE 1: Train head -----
    print("=" * 70)
    print("PHASE 1: TRAIN HEAD")
    print("=" * 70)

    compile_model(model, HEAD_LEARNING_RATE)
    cb1, ckpt1 = make_callbacks(run_name, run_dir, "phase1")

    t0 = time.time()
    h1 = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=PHASE1_EPOCHS,
        class_weight=cw,
        callbacks=cb1,
        verbose=1
    )
    t1 = time.time() - t0
    print(f"Phase 1: {t1 / 60:.2f} min")

    # ----- PHASE 2: Fine-tune -----
    print("=" * 70)
    print("PHASE 2: FINE-TUNE")
    print("=" * 70)

    unfreeze_top(base, FINE_TUNE_FRACTION)
    compile_model(model, FINETUNE_LEARNING_RATE)
    cb2, ckpt2 = make_callbacks(run_name, run_dir, "phase2")

    t0 = time.time()
    h2 = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=PHASE1_EPOCHS + PHASE2_EPOCHS,   # 15 + 35 = 50
        initial_epoch=len(h1.history["loss"]),
        class_weight=cw,
        callbacks=cb2,
        verbose=1
    )
    t2 = time.time() - t0
    print(f"Phase 2: {t2 / 60:.2f} min  |  Total: {(t1 + t2) / 60:.2f} min")

    model.save(os.path.join(run_dir, f"{run_name}.keras"))
    print("Checkpoints:", ckpt1, ckpt2)

    hist = merge_histories(h1, h2)

    for metric, ylabel in [("accuracy", "Accuracy"), ("loss", "Loss"), ("auc", "AUC")]:
        plot_history(
            hist.get(metric, []),
            hist.get(f"val_{metric}", []),
            f"{ylabel} - {run_name}",
            ylabel,
            os.path.join(run_dir, f"history_{metric}_{run_name}.png"),
        )

    result = evaluate_model(model, val_gen, test_dir, backbone_name, run_name, run_dir)
    result["Training Time (min)"] = (t1 + t2) / 60.0

    summary_df  = pd.DataFrame([result])
    summary_csv = os.path.join(run_dir, f"{run_name}_summary.csv")
    summary_df.to_csv(summary_csv, index=False)
    print("Saved:", summary_csv)

    try:
        display(HTML(summary_df.to_html(index=False)))
    except Exception:
        print(summary_df.to_string(index=False))

    del model, base, train_gen, val_gen, test_gen
    gc.collect()
    tf.keras.backend.clear_session()

    return summary_df


# ============================================================
# CELL 9: RUN BOTH MODELS SEQUENTIALLY + FINAL COMPARISON
# ============================================================
raw_base    = prepare_raw_dataset()
all_results = []

for backbone in BACKBONE_NAMES:
    print(f"\n{'='*90}")
    print(f"  STARTING MODEL: {backbone}")
    print(f"{'='*90}\n")
    result_df = run_raw_experiment(backbone, raw_base)
    all_results.append(result_df)

# ----- Final side-by-side comparison -----
final_comparison = pd.concat(all_results, ignore_index=True)

comparison_csv = os.path.join(OUTPUT_DIR, "final_comparison_both_models.csv")
final_comparison.to_csv(comparison_csv, index=False)
print("\n" + "=" * 90)
print("FINAL COMPARISON — MobileNetV2_S  vs  EfficientNetB3  (Raw Images)")
print("=" * 90)

try:
    display(HTML(final_comparison.to_html(index=False)))
except Exception:
    print(final_comparison.to_string(index=False))

print(f"\nComparison CSV saved to: {comparison_csv}")
```

```python
# ============================================================
# CELL 1: INSTALLS + IMPORTS
# ============================================================
import os
import sys
import gc
import cv2
import time
import glob
import shutil
import random
import zipfile
import warnings
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from pathlib import Path
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_curve,
    auc,
    roc_auc_score,
)
from sklearn.utils.class_weight import compute_class_weight
from google.colab import drive
from IPython.display import display, HTML

from tensorflow.keras import layers, models, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint

warnings.filterwarnings("ignore")

try:
    import cv2  # noqa
except ImportError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "opencv-python-headless"])
    import cv2

print("TensorFlow version:", tf.__version__)


# ============================================================
# CELL 2: GOOGLE DRIVE + CONFIG
# ============================================================
drive.mount("/content/drive", force_remount=False)

# -----------------------------
# DATA PATHS
# -----------------------------
RAW_ZIP_FILE_PATH = "/content/drive/MyDrive/1/isic2020.zip"

PROJECT_ROOT    = "/content/melanoma_project_raw_fair"
RAW_EXTRACT_DIR = os.path.join(PROJECT_ROOT, "raw_extracted")
OUTPUT_DIR      = os.path.join(PROJECT_ROOT, "outputs")

# -----------------------------
# GENERAL SETTINGS
# -----------------------------
SEED = 42

CLASSIFIER_INPUT_SHAPE = (224, 224, 3)
CLASSIFIER_IMG_SIZE    = CLASSIFIER_INPUT_SHAPE[:2]

DEFAULT_BATCH_SIZE  = 32

PHASE1_EPOCHS          = 15
PHASE2_EPOCHS          = 35
HEAD_LEARNING_RATE     = 1e-3
FINETUNE_LEARNING_RATE = 1e-5
LABEL_SMOOTHING        = 0.0
USE_CLASS_WEIGHTS      = True
FINE_TUNE_FRACTION     = 0.30

TTA_STEPS = 5

# -----------------------------
# BACKBONES TO RUN (Two models only)
# -----------------------------
BACKBONE_NAMES = ["MobileNetV2_S", "EfficientNetB3"]

# -----------------------------
# THRESHOLD TUNING
# -----------------------------
THRESHOLD_MIN = 0.10
THRESHOLD_MAX = 0.90
THRESHOLD_STEPS = 161
MIN_SPECIFICITY_FOR_THRESHOLD = 0.95

for p in [PROJECT_ROOT, RAW_EXTRACT_DIR, OUTPUT_DIR]:
    os.makedirs(p, exist_ok=True)

print("Project root :", PROJECT_ROOT)
print("Backbones    :", BACKBONE_NAMES)
print("Output dir   :", OUTPUT_DIR)


# ============================================================
# CELL 3: GENERAL UTILITIES
# ============================================================
def set_global_seed(seed=SEED):
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)

def ensure_dir(path):
    os.makedirs(path, exist_ok=True)

def unzip_dataset(zip_path, extract_to):
    if not os.path.exists(zip_path):
        raise FileNotFoundError(f"ZIP not found: {zip_path}")

    ensure_dir(extract_to)

    if len(os.listdir(extract_to)) > 0:
        print(f"Already extracted: {extract_to}")
        return True

    print(f"Extracting: {zip_path}")
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(extract_to)
    print(f"Extracted to: {extract_to}")
    return True

def find_base_dir(root_dir):
    if os.path.exists(os.path.join(root_dir, "train")):
        return root_dir
    for r, dirs, _ in os.walk(root_dir):
        if {"train", "val", "test"}.issubset(set(dirs)):
            return r
    return root_dir

def infer_positive_class(class_indices):
    inverse = {v: k for k, v in class_indices.items()}
    for idx, name in inverse.items():
        if "mel" in name.lower():
            return idx, name
    max_idx = max(inverse.keys())
    return max_idx, inverse[max_idx]

def get_negative_class_name(class_indices, positive_idx):
    inverse = {v: k for k, v in class_indices.items()}
    for idx, name in inverse.items():
        if idx != positive_idx:
            return name
    return "Negative"

set_global_seed(SEED)
print("Utilities ready.")


# ============================================================
# CELL 4: BACKBONE FACTORY + DATA GENERATORS
# ============================================================
def get_backbone(backbone_name):
    if backbone_name == "MobileNetV2_S":
        from tensorflow.keras.applications import MobileNetV2
        from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

        def fn(include_top=False, weights="imagenet", input_tensor=None, pooling="avg"):
            return MobileNetV2(
                include_top=include_top,
                weights=weights,
                input_tensor=input_tensor,
                pooling=pooling,
                alpha=0.75
            )
        return fn, preprocess_input, "lightweight"

    if backbone_name == "EfficientNetB3":
        from tensorflow.keras.applications import EfficientNetB3
        from tensorflow.keras.applications.efficientnet import preprocess_input

        def fn(include_top=False, weights="imagenet", input_tensor=None, pooling="avg"):
            return EfficientNetB3(
                include_top=include_top,
                weights=weights,
                input_tensor=input_tensor,
                pooling=pooling
            )
        return fn, preprocess_input, "stronger"

    raise ValueError(f"Unknown backbone: {backbone_name}")

def make_generators(base_dir, backbone_name):
    _, preprocess_input, _ = get_backbone(backbone_name)
    bs = DEFAULT_BATCH_SIZE

    def prep(img):
        return preprocess_input(img.astype(np.float32))

    train_gen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.03,
        height_shift_range=0.03,
        zoom_range=0.08,
        horizontal_flip=True,
        fill_mode="nearest",
        preprocessing_function=prep,
    ).flow_from_directory(
        os.path.join(base_dir, "train"),
        target_size=CLASSIFIER_IMG_SIZE,
        batch_size=bs,
        class_mode="binary",
        interpolation="bilinear",
        shuffle=True,
        seed=SEED,
    )

    val_test_dg = ImageDataGenerator(preprocessing_function=prep)

    val_gen = val_test_dg.flow_from_directory(
        os.path.join(base_dir, "val"),
        target_size=CLASSIFIER_IMG_SIZE,
        batch_size=bs,
        class_mode="binary",
        interpolation="bilinear",
        shuffle=False,
    )

    test_gen = val_test_dg.flow_from_directory(
        os.path.join(base_dir, "test"),
        target_size=CLASSIFIER_IMG_SIZE,
        batch_size=1,
        class_mode="binary",
        interpolation="bilinear",
        shuffle=False,
    )

    return train_gen, val_gen, test_gen

def get_class_weights(train_gen):
    y = train_gen.classes
    cw = compute_class_weight("balanced", classes=np.unique(y), y=y)
    d = {int(c): float(w) for c, w in zip(np.unique(y), cw)}
    print("Class weights:", d)
    return d


# ============================================================
# TTA PREDICTION FUNCTION
# ============================================================
def predict_with_tta(model, test_dir, backbone_name, tta_steps=TTA_STEPS):
    """
    Runs inference tta_steps times on the test set:
      - 1 pass with no augmentation (original)
      - (tta_steps - 1) passes with horizontal flip + small rotation + zoom
    Returns averaged probabilities, ground-truth labels, and class_indices.
    """
    _, preprocess_input, _ = get_backbone(backbone_name)

    def prep(img):
        return preprocess_input(img.astype(np.float32))

    # Pass 1: no augmentation
    base_dg = ImageDataGenerator(preprocessing_function=prep)
    base_gen = base_dg.flow_from_directory(
        test_dir,
        target_size=CLASSIFIER_IMG_SIZE,
        batch_size=1,
        class_mode="binary",
        interpolation="bilinear",
        shuffle=False,
    )
    print(f"TTA pass 1/{tta_steps} (original) ...")
    all_preds = model.predict(base_gen, steps=len(base_gen), verbose=1)

    # Passes 2..N: with augmentation
    tta_dg = ImageDataGenerator(
        horizontal_flip=True,
        rotation_range=10,
        zoom_range=0.05,
        fill_mode="nearest",
        preprocessing_function=prep,
    )

    for step in range(2, tta_steps + 1):
        tta_gen = tta_dg.flow_from_directory(
            test_dir,
            target_size=CLASSIFIER_IMG_SIZE,
            batch_size=1,
            class_mode="binary",
            interpolation="bilinear",
            shuffle=False,
        )
        print(f"TTA pass {step}/{tta_steps} (augmented) ...")
        all_preds += model.predict(tta_gen, steps=len(tta_gen), verbose=1)

    avg_preds = (all_preds / tta_steps).ravel()
    return avg_preds, base_gen.classes.astype(int), base_gen.class_indices


# ============================================================
# CELL 5: MODEL + TRAINING UTILITIES
# ============================================================
def build_model(backbone_name):
    bb_fn, _, profile = get_backbone(backbone_name)

    print("=" * 70)
    print(f"BUILDING: {backbone_name}  [{profile}]")
    print("=" * 70)

    inp = Input(shape=CLASSIFIER_INPUT_SHAPE, name="input_image")
    base = bb_fn(include_top=False, weights="imagenet", input_tensor=inp, pooling="avg")
    base.trainable = False

    x = layers.BatchNormalization(name="head_bn0")(base.output)
    x = layers.Dropout(0.35, name="head_do0")(x)

    x = layers.Dense(512, name="head_fc1")(x)
    x = layers.BatchNormalization(name="head_fc1_bn")(x)
    x = layers.LeakyReLU(negative_slope=0.2, name="head_fc1_act")(x)
    x = layers.Dropout(0.40, name="head_do1")(x)

    x = layers.Dense(256, name="head_fc2")(x)
    x = layers.BatchNormalization(name="head_fc2_bn")(x)
    x = layers.LeakyReLU(negative_slope=0.2, name="head_fc2_act")(x)
    x = layers.Dropout(0.30, name="head_do2")(x)

    out = layers.Dense(1, activation="sigmoid", name="classifier")(x)

    model = models.Model(inputs=inp, outputs=out, name=f"{backbone_name}_Transfer")

    print(f"Total params:    {model.count_params():,}")
    print(f"Backbone params: {base.count_params():,}")
    return model, base

def compile_model(model, lr):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss=tf.keras.losses.BinaryCrossentropy(label_smoothing=LABEL_SMOOTHING),
        metrics=[
            tf.keras.metrics.BinaryAccuracy(name="accuracy"),
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc"),
        ],
    )

def unfreeze_top(base, fraction=FINE_TUNE_FRACTION):
    base.trainable = True
    n = len(base.layers)
    cut = int(n * (1.0 - fraction))

    for i, layer in enumerate(base.layers):
        layer.trainable = i >= cut and not isinstance(layer, layers.BatchNormalization)

    print(f"Fine-tuning top {fraction * 100:.0f}% ({n - cut}/{n} layers)")

def make_callbacks(run_name, run_dir, phase):
    ckpt = os.path.join(run_dir, f"best_{run_name}_{phase}.keras")
    return [
        ModelCheckpoint(ckpt, monitor="val_auc", mode="max", save_best_only=True, verbose=1),
        ReduceLROnPlateau(
            monitor="val_auc",
            mode="max",
            factor=0.5,
            patience=4,
            min_lr=1e-7,
            verbose=1
        ),
        EarlyStopping(
            monitor="val_auc",
            mode="max",
            patience=8,
            min_delta=1e-4,
            restore_best_weights=True,
            verbose=1
        ),
    ], ckpt

def merge_histories(h1, h2):
    keys = sorted(set(list(h1.history) + list(h2.history)))
    return {k: h1.history.get(k, []) + h2.history.get(k, []) for k in keys}

def plot_history(train_vals, val_vals, title, ylabel, save_path):
    if not train_vals and not val_vals:
        return

    plt.figure(figsize=(8, 6))
    if train_vals:
        plt.plot(range(1, len(train_vals) + 1), train_vals, label=f"Train {ylabel}", linewidth=2)
    if val_vals:
        plt.plot(range(1, len(val_vals) + 1), val_vals, label=f"Val {ylabel}", linewidth=2)

    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()


# ============================================================
# CELL 6: THRESHOLD TUNING + EVALUATION
# ============================================================
def plot_cm(cm, labels, title, path):
    plt.figure(figsize=(8, 6))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=labels,
        yticklabels=labels,
        cbar=False,
        annot_kws={"size": 14},
    )
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()

def plot_roc(y_true, y_score, title, path, pos_name="Positive"):
    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, lw=2, label=f"{pos_name} (AUC = {roc_auc:.4f})")
    plt.plot([0, 1], [0, 1], "--", lw=1)
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.2)
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()

def choose_best_threshold(y_true, y_prob, min_specificity=0.95):
    best_t = 0.5
    best_score = -1.0
    rows = []

    thresholds = np.linspace(THRESHOLD_MIN, THRESHOLD_MAX, THRESHOLD_STEPS)

    for t in thresholds:
        y_pred = (y_prob >= t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred, labels=[0, 1]).ravel()

        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0.0
        recall      = recall_score(y_true, y_pred, zero_division=0)
        precision   = precision_score(y_true, y_pred, zero_division=0)
        f1          = f1_score(y_true, y_pred, zero_division=0)

        if min_specificity is not None and specificity < min_specificity:
            continue

        rows.append({
            "threshold":   t,
            "specificity": specificity,
            "recall":      recall,
            "precision":   precision,
            "f1":          f1
        })

        if f1 > best_score:
            best_score = f1
            best_t = t

    df = pd.DataFrame(rows)

    if df.empty:
        print("No threshold satisfied the specificity constraint. Falling back to 0.5")
        return 0.5, df

    return float(best_t), df

def normalize_positive_class(y_true, y_prob, class_indices):
    pos_idx, pos_name = infer_positive_class(class_indices)
    neg_name = get_negative_class_name(class_indices, pos_idx)

    y_true = y_true.astype(int).copy()
    y_prob  = y_prob.copy()

    if pos_idx != 1:
        y_true = 1 - y_true
        y_prob  = 1.0 - y_prob

    return y_true, y_prob, pos_name, neg_name

def evaluate_model(model, val_gen, test_dir, backbone_name, run_name, run_dir):
    print("=" * 70)
    print(f"EVALUATION: {run_name}")
    print("=" * 70)

    # --- Validation probs for threshold tuning (no TTA on val) ---
    val_probs = model.predict(val_gen, steps=len(val_gen), verbose=1).ravel()
    y_val     = val_gen.classes.astype(int)

    y_val, val_probs, pos_name, neg_name = normalize_positive_class(
        y_val, val_probs, val_gen.class_indices
    )

    best_thresh, thr_df = choose_best_threshold(
        y_val, val_probs, min_specificity=MIN_SPECIFICITY_FOR_THRESHOLD
    )

    thr_csv = os.path.join(run_dir, f"threshold_search_{run_name}.csv")
    thr_df.to_csv(thr_csv, index=False)
    print(f"Chosen threshold: {best_thresh:.4f}")
    print("Saved threshold table:", thr_csv)

    # --- Test probs with TTA ---
    t0 = time.time()
    probs, y_true, class_indices = predict_with_tta(model, test_dir, backbone_name, TTA_STEPS)
    elapsed = time.time() - t0
    per_img = elapsed / max(len(probs), 1)

    y_true, probs, pos_name, neg_name = normalize_positive_class(
        y_true, probs, class_indices
    )

    y_pred = (probs >= best_thresh).astype(int)

    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()

    specificity = (tn / (tn + fp) * 100) if (tn + fp) > 0 else 0.0
    roc_auc     = roc_auc_score(y_true, probs)

    plot_roc(
        y_true, probs,
        f"ROC Curve - {run_name}",
        os.path.join(run_dir, f"roc_{run_name}.png"),
        pos_name=pos_name
    )

    plot_cm(
        cm,
        [neg_name, pos_name],
        f"Confusion Matrix - {run_name}",
        os.path.join(run_dir, f"cm_{run_name}.png")
    )

    np.savez(
        os.path.join(run_dir, f"predictions_{run_name}.npz"),
        y_true=y_true, y_pred=y_pred, y_prob=probs, threshold=best_thresh,
    )

    acc  = accuracy_score(y_true, y_pred) * 100
    pre  = precision_score(y_true, y_pred, average="weighted", zero_division=0) * 100
    rec  = recall_score(y_true, y_pred, average="weighted", zero_division=0) * 100
    f1w  = f1_score(y_true, y_pred, average="weighted", zero_division=0) * 100

    p_pr = precision_score(y_true, y_pred, pos_label=1, average="binary", zero_division=0) * 100
    p_re = recall_score(y_true, y_pred, pos_label=1, average="binary", zero_division=0) * 100
    p_f1 = f1_score(y_true, y_pred, pos_label=1, average="binary", zero_division=0) * 100

    mdf = pd.DataFrame({
        "Metric": [
            "Chosen Threshold", "TTA Steps",
            "Accuracy (ACC)", "Precision Weighted", "Recall Weighted",
            "Specificity (SPE)", "F1 Weighted",
            f"{pos_name} Precision", f"{pos_name} Recall", f"{pos_name} F1",
            "ROC AUC", "Test Set Time", "Per Image Time",
        ],
        "Value": [
            f"{best_thresh:.4f}", f"{TTA_STEPS}",
            f"{acc:.2f}%", f"{pre:.2f}%", f"{rec:.2f}%",
            f"{specificity:.2f}%", f"{f1w:.2f}%",
            f"{p_pr:.2f}%", f"{p_re:.2f}%", f"{p_f1:.2f}%",
            f"{roc_auc:.4f}", f"{elapsed:.2f} s", f"{per_img:.4f} s",
        ],
    })

    csv_path = os.path.join(run_dir, f"metrics_{run_name}.csv")
    mdf.to_csv(csv_path, index=False)
    print("Saved:", csv_path)

    try:
        display(HTML(mdf.to_html(index=False)))
    except Exception:
        print(mdf.to_string(index=False))

    return {
        "Run": run_name, "Backbone": backbone_name, "TTA Steps": TTA_STEPS,
        "ACC": acc, "PRE": pre, "REC": rec, "SPE": specificity, "F1 Score": f1w,
        f"{pos_name} Precision": p_pr, f"{pos_name} Recall": p_re, f"{pos_name} F1": p_f1,
        "ROC AUC": roc_auc, "Threshold": best_thresh,
        "Execution Time (s)": elapsed, "Per Image Test Time (s)": per_img,
    }


# ============================================================
# CELL 7: PREPARE RAW DATASET
# ============================================================
def prepare_raw_dataset():
    ensure_dir(PROJECT_ROOT)
    ensure_dir(OUTPUT_DIR)

    unzip_dataset(RAW_ZIP_FILE_PATH, RAW_EXTRACT_DIR)

    raw_base = find_base_dir(RAW_EXTRACT_DIR)
    print("Raw base:", raw_base)

    if not os.path.exists(os.path.join(raw_base, "train")):
        raise FileNotFoundError("train/val/test not found in raw dataset.")

    return raw_base


# ============================================================
# CELL 8: RUN EXPERIMENT FOR ONE BACKBONE
# ============================================================
def run_raw_experiment(backbone_name, base_dir):
    tf.keras.backend.clear_session()
    gc.collect()
    set_global_seed(SEED)

    safe     = "raw_fair"
    run_name = f"{backbone_name}_{safe}"
    run_dir  = os.path.join(OUTPUT_DIR, backbone_name, safe)
    ensure_dir(run_dir)

    test_dir = os.path.join(base_dir, "test")

    print("\n" + "#" * 90)
    print(f"  RUNNING: {run_name}")
    print("#" * 90)
    print("Data dir:", base_dir)

    train_gen, val_gen, test_gen = make_generators(base_dir, backbone_name)

    print(f"Train: {train_gen.samples}  Val: {val_gen.samples}  Test: {test_gen.samples}")
    print("Classes:", train_gen.class_indices)
    print(f"TTA steps: {TTA_STEPS}")

    cw = get_class_weights(train_gen) if USE_CLASS_WEIGHTS else None

    model, base = build_model(backbone_name)

    # ----- PHASE 1: Train head -----
    print("=" * 70)
    print("PHASE 1: TRAIN HEAD")
    print("=" * 70)

    compile_model(model, HEAD_LEARNING_RATE)
    cb1, ckpt1 = make_callbacks(run_name, run_dir, "phase1")

    t0 = time.time()
    h1 = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=PHASE1_EPOCHS,
        class_weight=cw,
        callbacks=cb1,
        verbose=1
    )
    t1 = time.time() - t0
    print(f"Phase 1: {t1 / 60:.2f} min")

    # ----- PHASE 2: Fine-tune -----
    print("=" * 70)
    print("PHASE 2: FINE-TUNE")
    print("=" * 70)

    unfreeze_top(base, FINE_TUNE_FRACTION)
    compile_model(model, FINETUNE_LEARNING_RATE)
    cb2, ckpt2 = make_callbacks(run_name, run_dir, "phase2")

    t0 = time.time()
    h2 = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=PHASE1_EPOCHS + PHASE2_EPOCHS,   # 15 + 35 = 50
        initial_epoch=len(h1.history["loss"]),
        class_weight=cw,
        callbacks=cb2,
        verbose=1
    )
    t2 = time.time() - t0
    print(f"Phase 2: {t2 / 60:.2f} min  |  Total: {(t1 + t2) / 60:.2f} min")

    model.save(os.path.join(run_dir, f"{run_name}.keras"))
    print("Checkpoints:", ckpt1, ckpt2)

    hist = merge_histories(h1, h2)

    for metric, ylabel in [("accuracy", "Accuracy"), ("loss", "Loss"), ("auc", "AUC")]:
        plot_history(
            hist.get(metric, []),
            hist.get(f"val_{metric}", []),
            f"{ylabel} - {run_name}",
            ylabel,
            os.path.join(run_dir, f"history_{metric}_{run_name}.png"),
        )

    result = evaluate_model(model, val_gen, test_dir, backbone_name, run_name, run_dir)
    result["Training Time (min)"] = (t1 + t2) / 60.0

    summary_df  = pd.DataFrame([result])
    summary_csv = os.path.join(run_dir, f"{run_name}_summary.csv")
    summary_df.to_csv(summary_csv, index=False)
    print("Saved:", summary_csv)

    try:
        display(HTML(summary_df.to_html(index=False)))
    except Exception:
        print(summary_df.to_string(index=False))

    del model, base, train_gen, val_gen, test_gen
    gc.collect()
    tf.keras.backend.clear_session()

    return summary_df


# ============================================================
# CELL 9: RUN BOTH MODELS SEQUENTIALLY + FINAL COMPARISON
# ============================================================
raw_base    = prepare_raw_dataset()
all_results = []

for backbone in BACKBONE_NAMES:
    print(f"\n{'='*90}")
    print(f"  STARTING MODEL: {backbone}")
    print(f"{'='*90}\n")
    result_df = run_raw_experiment(backbone, raw_base)
    all_results.append(result_df)

# ----- Final side-by-side comparison -----
final_comparison = pd.concat(all_results, ignore_index=True)

comparison_csv = os.path.join(OUTPUT_DIR, "final_comparison_both_models.csv")
final_comparison.to_csv(comparison_csv, index=False)
print("\n" + "=" * 90)
print("FINAL COMPARISON — MobileNetV2_S  vs  EfficientNetB3  (Raw Images)")
print("=" * 90)

try:
    display(HTML(final_comparison.to_html(index=False)))
except Exception:
    print(final_comparison.to_string(index=False))

print(f"\nComparison CSV saved to: {comparison_csv}")
```
